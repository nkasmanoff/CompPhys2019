# b) Modify your program to calculate the Fourier transform of the sunspot data and then make a graph of the magnitude squared |ck|2 of the Fourier coefficients as a function of k—also called the power spectrum of the sunspot signal. You should see that there is a noticeable peak in the power spectrum at a nonzero value of k. The appearance of this peak tells us that there is one frequency in the Fourier series that has a higher amplitude than the others around it—meaning that there is a large sine-wave term with this frequency, which corresponds to the periodic wave you can see in the original data.


```python
def dft(y):
    """
    Code to calculate the discrete fourier transform c_k.
    
    Note that there will both a real and imaginary value for every c, 
    
    so to find the maximum k we need to calculate the power spectrum .
    
    """
    N = len(y) #we create the interval as the length of the dataset.. 
    
    c = np.zeros(N,complex) #list of fourier coef.

    for k in range(N): # Iterate over possible k's
        for n in range(N): #Iterate over all n's
            c[k] += y[n]*np.exp(-2j*np.pi*k*n/N) #create a sum to solve for c[k]
    return c

c = dft(y)
```

```python
power = np.real(c)**2 + np.imag(c)**2
plt.plot(power) #k is just the index number. 

plt.ylabel("$|c(k)|^2$",fontsize=20)
plt.xlabel('$k$',fontsize=20)
```

Immediately from this graph we observe the nyquist frequency at the end, and can ignore that. Furthermore, we see that the peak in $k$ is somewhere very early on in the first interval, and can zoom in on this figure to better capture this value. Our original guess for the wavelength was 130 months, and if we apply the formula $\lambda = \frac{N}{k_{max}} = 130.95$, which is almost exactly what we guessed! While this dataset of sunspot data was relatively clean and it was easy to infer the periodicity, this is far from true and reality and fourier analysis via the discrete fourier transform proves to be an invaluable (but computationally expensive) tool. To remedy this last issue, we will demonstrate the power of the FFT in the following next section.

```python
def idft(c,find_k=False):
    """
    Inverse Discrete Fourier Transform (idf) 
    """
    N = len(c) #we create the interval as the length of the dataset.. 
    y = np.zeros(N,complex) #The number of y's to recover. 
    if find_k != False:
        k = find_k
        for n in range(N):
            y[n] += c[k]*np.exp(2j*np.pi*k*n/N)/N
        
        
        return y
    #of the interval. 
    
    for n in range(N):
        for k in range(N): #only need to compute sine or cosine terms, 
            y[n] += c[k]*np.exp(2j*np.pi*k*n/N)/N
    return y

y_kmax = idft(c,find_k=k_max)
y_recovered =  idft(c)
```

```python
plt.figure(figsize=(8,8))
plt.plot(y,'k.',label='data')
plt.plot(np.real(y_kmax),label = 'Peak Wave #')
plt.plot(y_recovered,'r',label = "Full IDFT")
plt.legend()
```
